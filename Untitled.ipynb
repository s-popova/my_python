```python
class Animal:
    def __init__(self, name='Животное без названия',x=0, y=0):
        
        self.name = name
        self._item = 0 
        self.x = x
        self.y = y
        print(f"Инициализация координат ({x},{y})")
    def __str__(self):
        return f'{self.name}' 
        if self._item == 0: 
            print("Надо собрать предмет!!!")
    def shoot(self):
        print(f"{self.name} стреляет")
    def running(self):
        if self._item == 0:
            print(f'у {self.name} Нет сил бежать...')   
        else:
            self._item -= 1
            self.x+=1
            print(f"{self.name} бежит и теряет предмет")
            print(f'координаты ({self.x},{self.y})')
    def collecting(self):
        self._item += 1 
        print(f'{self.name} собирает предметы')
        print(f'всего предметов у {self.name}: ({self._item})')

class Bird(Animal):
    def __init__(self, name='Птица без названия', x=0, y=0):
        Animal.__init__(self, name=name,x=x, y=y)
        self.fly = ' умеет летать'
    def __str__(self):
        return f'{self.name} {self.fly}'                
    def set_fly(self, fly):
        self.fly = fly   
    def running(self):
        if self._item == 0:
            print(f'у {self.name} нет сил лететь ...')   
        else:
            print("птичка летит и теряет предмет")
            self._item -= 1 
            self.x+=1
            self.y+=1
            print(f'координаты {self.name}: ({self.x},{self.y})')
      

my_animal = Animal('Лев', 1 , 0)
print(my_animal)

my_bird = Bird('Грач',6,9)
print(my_bird)

my_animal1 = Animal('Панда', 1 , 8)
print(my_animal1)
my_animal.running()
my_animal.shoot()
my_animal.collecting()
my_animal.running()
my_bird.shoot()
my_bird.collecting()
my_bird.running()
my_animal1.collecting()
my_animal1.collecting()
my_animal1.collecting()
my_animal1.running()
```
